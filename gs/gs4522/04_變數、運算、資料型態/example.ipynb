# Arithmetic Operators

```python
a = 5
b = 3

sum_num = a + b
sub_num = a - b
mul_num = a * b
div_num = a / b
mod_num = a % b
exp_num = a ** b
int_div_num = a // b

print('a + b = {}'.format(sum_num))
print('a - b = {}'.format(sub_num))
print('a * b = {}'.format(mul_num))
print('a / b = {}'.format(div_num))
print('a % b = {}'.format(mod_num))
print('a ** b = {}'.format(exp_num))
print('a // b = {}'.format(int_div_num))
```
