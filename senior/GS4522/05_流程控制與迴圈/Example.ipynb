# while loop

```python
i = 1
while i < 6:
    print(i)
    i += 1 # i = i + 1
```

```python
i = 0
num_sum = 0
while i <= 100:
    num_sum = num_sum + i
    i = i + 1
    #print('num_sum: {}, i: {}'.format(num_sum, i))
print('1+2+......+100 = {}'.format(num_sum))
```

```python
i = 0
num_sum = 0
while i <= 100:
    if i % 2 == 1:
        i = i + 1
        continue
    num_sum = num_sum + i
    print('num_sum: {}, i: {}'.format(num_sum, i))

    i = i + 1
print('2+4+6......+100 = {}'.format(num_sum))
```

```python
i = 0
num_sum = 0
while True:
    if i > 100:
        break
    num_sum = num_sum + i
    i = i + 1
print('1+2+......+100 = {}'.format(num_sum))
```

# nested loop

```python
n=5
for i in range(n):
    for j in range(i):
#        print('i: {} , j: {}'.format(i,j))
        print ('* ', end="")
    print('')

for i in range(n,0,-1):
    for j in range(i):
        print('* ', end="")
    print('')
```

# list comprehension

```python
x = [i for i in range(10)]
print(x)
```

```python
x = [i for i in range(50) if i % 5 ==0]
print(x)
```

```python
answer = 3
guess = 0

while guess!=answer:
    guess = int(input('please make guess during 1~6: '))
    if guess > answer:
        print('bigger than answer')
    elif guess < answer:
        print('smaller than answer')
    else:
        print('bingo')
```
