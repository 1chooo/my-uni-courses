```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from getmac import get_mac_address as gma
mac = gma()

fig = plt.figure()
ax = Axes3D(fig)
def z_func(x, y):
    return ( (-20) * np.exp((-0.2) * ((x**2 + y**2) * 0.5)**(0.5)) ) - \
           np.exp( (0.5) * (np.cos(2*np.pi*x) + np.cos(2*np.pi*y) ) ) + \
            np.exp(1) + 20
x = y = np.arange(-6, 6, 0.1) #-2 to 2.1 0.2
X, Y = np.meshgrid(x, y) #
Z = z_func(X, Y)
ec = ax.plot_surface(X, Y, Z, cmap='plasma') #X, Y:
                                             #cmap:plot_surfaceex:plasma, coolwarm
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z', fontsize=10, rotation=0)

cbar = plt.colorbar(ec, orientation='horizontal', shrink=0.5)# colorbar()or, 
cbar.set_label('Z')

plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='109601005 '+mac) #
fig.savefig('fex_2_1.png', dpi=300)
plt.show()





fig = plt.figure(figsize=(6, 6))
CS = plt.contour(X, Y, Z, 20)
cbar = plt.colorbar(CS, orientation = 'horizontal', shrink =0.7, )
plt.xlabel('X')
plt.ylabel('Y')
plt.title('109601005 '+mac)
fig.savefig('fex_2_2.png', dpi=300)
plt.show()
```
